```python
#Importing the libraries
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the Dataset
data = pd.read_csv('bestsellers with categories.csv')
```

```python
#printing the first five line
data.head()
```

```python
#checking for missing values
data.isnull().sum()
```

```python
#look for correlation matrix to check which columns are correlated with eachother
corr=data.corr()
```

```python
sns.heatmap(corr)
```

```python
#lets do some visualization
#1 TOP TEN SELLING BOOK OF THE DECADE
data[['Name','Year']].nlargest(10,'Year').plot.bar(x='Name',y='Year',color='purple')
plt.title('Top 10 seeling books of the decade',size=15)
plt.xlabel('Name of the Books',fontsize=13)
plt.ylabel('Price')
plt.show()
```

```python
#2 THE PERCENTAGE OF FICTION TO NON FICTION
data['Genre'].value_counts()
```

```python
slices=[310,240]
activities=['Non Fiction', 'Fiction']
colors=['b','g']
plt.pie(slices, labels=activities, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('Percentage of each genre')
plt.show()
```

```python
#3 TREND OVER THE YEARS
data[['Year','User Rating']].nlargest(20,'User Rating').plot.bar(x='Year',y='User Rating',color='blue')
plt.title('Trend over the years',size=8)
plt.xlabel('Name of the Books',fontsize=7)
plt.ylabel('Rating')
plt.show()
```

```python
#4 TOP TEN AUTHORS ACCORDING TO REVIEWS
data[['Author','Reviews']].nlargest(10,'Reviews').plot.bar(x='Author',y='Reviews',color='Yellow')
plt.title('Top 10 Author according to reviews',size=15)
plt.xlabel('Authors',fontsize=13)
plt.ylabel('Numbers of Reviews')
plt.show()
```
